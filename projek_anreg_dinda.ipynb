```python

# ===============================
# 1. IMPORT LIBRARY
# ===============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# ===============================
# 2. LOAD DATA
# ===============================
data = pd.read_csv("female-and-male-life-expectancy-at-birth-in-years.csv")
print(data.columns.tolist())

# Bersihkan nama kolom
data.columns = data.columns.str.strip()

print("Kolom Data:")
print(data.columns)
print(data.head())

# ===============================
# 3. FILTER INDONESIA (OPSIONAL)
# ===============================
data = data[data["Entity"] == "Indonesia"]
print(data)

# ===============================
# 4. HAPUS MISSING VALUE
# ===============================
data = data.dropna()

# ===============================
# 5. ANALISIS DESKRIPTIF
# ===============================
print("\nStatistik Deskriptif:")
print(data[["Men", "Women"]].describe())

# ===============================
# 6. VISUALISASI
# ===============================

# Grafik Tren
plt.figure(figsize=(10,6))
plt.plot(data["Year"], data["Men"], label="Men")
plt.plot(data["Year"], data["Women"], label="Women")
plt.title("Life Expectancy Indonesia")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.show()

# Boxplot (Deteksi Outlier Visual)
plt.figure(figsize=(8,5))
sns.boxplot(data=data[["Men", "Women"]])
plt.title("Deteksi Outlier (Boxplot)")
plt.show()

# ===============================
# 7. KORELASI YEAR vs MAN
# ===============================
corr_men, p_men = pearsonr(data["Year"], data["Men"])

print("\nKorelasi Year vs Men:")
print("Koefisien:", corr_men)
print("P-value:", p_men)

# ===============================
# 8. KORELASI YEAR vs WOMAN
# ===============================
corr_women, p_women = pearsonr(data["Year"], data["Women"])

print("\nKorelasi Year vs Women:")
print("Koefisien:", corr_women)
print("P-value:", p_women)

print("\nAnalisis Selesai ✅")
```
